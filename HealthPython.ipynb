```python
df = pd.read_csv('modified_healthcare_dataset.csv')
```

```python
df
```

```python
df_practice = df.copy()
df_practice.columns = df_practice.columns.str.upper()
display(df_practice)
```

```python
df_practice = df_practice.drop_duplicates()
df_practice
```

```python
missing_values = df_practice.isnull().sum()
display(missing_values)
```

```python
df_practice['DATE OF ADMISSION'] = pd.to_datetime(df_practice['DATE OF ADMISSION'])
df_practice['DISCHARGE DATE'] = pd.to_datetime(df_practice['DISCHARGE DATE'])
print("Converted 'DATE OF ADMISSION' and 'DISCHARGE DATE' columns to datetime objects.")
```

```python
df_practice['ADMISSION_YEAR'] = df_practice['DATE OF ADMISSION'].dt.year
df_practice['ADMISSION_MONTH'] = df_practice['DATE OF ADMISSION'].dt.month
print("Created 'ADMISSION_YEAR' and 'ADMISSION_MONTH' columns.")
```

```python
df_practice['ADMISSION_DAY_OF_WEEK'] = df_practice['DATE OF ADMISSION'].dt.day_name()
print("Created 'ADMISSION_DAY_OF_WEEK' column.")
```

```python
df_practice['CALCULATED_LENGTH_OF_STAY'] = (df_practice['DISCHARGE DATE'] - df_practice['DATE OF ADMISSION']).dt.days
print("Created 'CALCULATED_LENGTH_OF_STAY' column.")
```
